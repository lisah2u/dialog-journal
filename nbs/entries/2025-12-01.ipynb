During the fourth of five weeks of Jeremy and John's [solveit](https://solve.it.com) course, I realized that my digital notebook ([capacities](https://capacities.io) app, for me) was overflowing with inter-mingled content that is a mix of code and notes about solveit, nbdev, jupyter, quarto, ssage, tmux, etc. Having been an avid digital note-taker for many years, I'm fairly disciplined about capturing and organizing notes on complex topics. I not only use capacities to organize raw notes, I also use [Heptabase](https://heptabase.com) for visual white-boarding of topics, ideas, and books where the focus is organizing and refining my thinking. 

Through this process, I realized that I now have a new sort of artifact to acount for: dialogs.  

Thus, I spent the weekend thinking about how to organize what I'm learning via this new media. Not only will I forget what I've learned, if I don't wrangle this quickly -- I will be lost in a sea of what-was-that-interesting-dialog-that related-to-the-thing-I'm-doing-now. Moreover, there are many interesting, community-shared dialogs that I'd like to go back and study. Thus, while this minimal nbdev notebook still lacks much structure. It presents a start towards organizing media, thoughts, and ideas for improvement.

There are also a bunch of interesting ideas I have about how to enrich dialogs inside of solveit. For example, task-based dialogues are structured. But not necessarily as a pre-built tree, rather as unfolding based on constraints of linguistic structure, intentional, and attentional. (I haven't read Grosz and Sidner's paper on [attention, intention, and the structure of discourse](https://aclanthology.org/J86-3001.pdf) in many years - so perhaps its time for a re-read in this age of LLMs.) Regardless, the human using solveit ends up having to post-edit dialogs in order to better structure them for reading. This is fine, and actually helpful, since through re-reading we have an opportunity to better organize thoughts and learn. But it would also be great to add features more geared towards reading and remembering.

For example, one small idea would be to focus some attention towards improving the table-of-contents and sidebar. Recently, Jeremy added a great little feature -- bookmarks -- which draws attention to some specific cell in the dialog that the author would like to call out via a navigational link. What about also colorizing TOC levels to make it easier to see the structure of conversation as organized by the author? 

Another small idea would be to make it easier to "fork" a sub-dialogue to a new conversation. Or to add an annotation pane for meta discourse about parts of the dialog. Well... there are tons of ideas and I'm really having fun thinking about where this is going!

Of particular interest to me right now is the need to become more proficient at using the solveit tool. The method for writing code write a bit, then verify output -- in small iterative loops, is delightful. I am also really enjoying the [fast.ai coding style](https://docs.fast.ai/dev/style.html) since I haven't been coding in my job for a long time. That said, one of my biggest take-aways of this class is that becoming proficient with the tool via learning shortcuts is essential. The better you can manipulate both the UI and manage the context for the LLLM, the better. Thus, while I meant to create a small tool this past weekend for spaced repetition of solveit keyboard shortcut practice in generated notebook... I ended up setting this journal instead. That project can be a nice follow-on to this.
