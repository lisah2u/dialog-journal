A Jupyter notebook is just a JSON file. You can load it with Python's `json` module or use `nbformat`, which is designed for this. The basic structure is:

```python
import nbformat

# Load notebook
nb = nbformat.read('path/to/notebook.ipynb', as_version=4)

# Save it back

```
